Dataset used: Financial Budget 2021

Importing the necessary Python libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Reading the CSV file

```python
data = pd.read_csv("Department Ministry,Fund allotted(i.csv")
data.head()
```

Checking if there are any missing values in the dataset

```python
data.isna()
```

Removing the null values.

```python
data.dropna()
```

Combining the smaller departments into the main ones.

```python
data = data.iloc[[0,8,11,14,18,23,41,42,43],:]
row = {'Department /Ministry': 'OTHERS', 'Fund allotted(in ₹crores)': 592971.0800000001}
data = data.append(row, ignore_index = True)
print(data)
```

Plotting a bar graph of the funds alloted to the various ministries.

```python
data.plot.bar(x='Department /Ministry', y='Fund allotted(in ₹crores)')
```

Plotting a donut chart of the funds allotted to the various ministries.

```python
df = data["Fund allotted(in ₹crores)"]
labels = data["Department /Ministry"]
plt.figure(figsize=(7,7))
plt.pie(df, labels=labels, autopct='%1.1f%%', startangle=90, pctdistance=0.85, shadow =True)
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Distribution of The Budget", fontsize=20)
plt.show()
```

Thus, we can see that the Ministry of Finance was allotted the most funds, 40% of the total 2021 budget.
